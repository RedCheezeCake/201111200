```python
import urllib
from lxml.cssselect import CSSSelector
import lxml.html

response= urllib.urlopen('https://www.google.com/finance/historical?q=KRX%3AKOSPI200&ei=wI_0V4nxEcWS0AS3hYCQAw')
googleFinance = response.read()

ghtml = lxml.html.fromstring(googleFinance)

for i in range(31):
    for j in range(5):
        sel1 = CSSSelector('#prices > table > tr:nth-child('+str(i+1)+') > td:nth-child('+str(j+1)+')')
        result = sel1(ghtml)
        for r in result:
            print r.text
    sel2 = CSSSelector('#prices > table > tr:nth-child('+str(i+1)+') > td.rgt.rm')
    result = sel2(ghtml)
    for r in result:
        print r.text
```
